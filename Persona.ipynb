```python
#i decided to establish segmentation as "15_18","18_30","30_45","45_55","55_65".
bins=[15,18,30,45,55,65]
labels=["15_18","18_30","30_45","45_55","55_65"]
```

```python
df["AGE"].value_counts()
```

```python
agg_df["AGE_CAT"]=pd.cut(df["AGE"],bins=bins,labels=labels)
```

```python
agg_df["AGE_CAT"].value_counts()
```

```python
agg_df.head()
```

```python
df.index.equals(agg_df.index) #i control index counts to ensure that there is no loss of data.
```

```python
#6

agg_df["customers_level_based"]=agg_df[["COUNTRY","SOURCE","SEX","AGE_CAT"]].apply(lambda x:"_".join(map(str,x)).upper(),axis=1)
```

```python
agg_df.head()
```

```python
#7 Also i create segment according to price. Weather "D" is low "A" is high.
agg_df["SEGMENT"] = pd.cut(agg_df["PRICE"], 4, labels=["D", "C", "B", "A"])
```

```python
agg_df[agg_df["AGE_CAT"]=="15_18"].groupby("SEGMENT").agg({"PRICE":["mean","max","sum"]})
```

```python
#8 According to customer segmentation that i created earlier, i made pivot table that show price segmentation of customer.
#Customer is Turkish women who uses android is between 30 and 45 years old.
new_user="TUR_ANDROID_FEMALE_30_45"
agg_df[agg_df["customers_level_based"]==new_user].groupby("SEGMENT").agg({"PRICE":"mean"})
```

```python
#Also i tried  segmentation method for Ios user french women who is between 30 to 45 years old.
new_user2="FRA_IOS_FEMALE_30_45"
agg_df[agg_df["customers_level_based"]==new_user2].groupby("SEGMENT").agg({"PRICE":"mean"})
```
